```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
prob = 1-np.cumprod(np.array(np.arange(1,366)/365)[::-1])
x = np.arange(1, 101)
plt.plot(x+1, prob[x])
plt.axhline(0.5, color="red", linestyle='--')
plt.axvline(23, color="green", linestyle=":")
plt.xticks(np.arange(0, 110, 10))
plt.yticks(np.linspace(0, 1, 11))
plt.xlabel("Group Size")
plt.ylabel("Probability of having same birthday")
plt.title("Birthday Problem")
plt.savefig("birthday_problem.png")
plt.show()
```
